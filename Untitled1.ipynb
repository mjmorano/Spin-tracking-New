```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gpuRes = [1,3357,
		10,3342,
		100,3347,
		1000,3525,
		2000,3480,
		3000,3475,
		4000,3474,
		5000,3476,
		6000,3473,
		7000,3471,
		8000,3495,
		9000,3473,
		10000,3469,
		11000,5330,
		12000,5486,
		13000,5795,
		14000,6307,
		15000,7538,
		16000,8531,
		17000,8753,
		18000,8798,
		19000,8947,
		20000,9271]

cpuRes = [1,583,
		10,682,
		100,1827,
		1000,18941,
		2000,41291,
		3000,57722,
		4000,81126,
		5000,106318,
		6000,128647,
		7000,123610,
		8000,141399]

gpuRes = np.array(gpuRes).reshape((-1, 2)).T
cpuRes = np.array(cpuRes).reshape((-1, 2)).T
```

```python
plt.plot(gpuRes[0,:], gpuRes[1,:]/1000, label='Nvidia V100')
plt.plot(cpuRes[0,:], cpuRes[1,:]/1000, label='x4 Cores of Xeon Gold 6428')
plt.yscale('log')
plt.title('Comparing CPU and GPU Performance for a 1 second simulation')
plt.legend()
plt.xlabel("Number of Particles at Once")
plt.ylabel("Time Taken (s)")
plt.show()
```

```python
averageCPUPerf = np.mean(cpuRes[1,3:]/cpuRes[0,3:])
```

```python
plt.plot(gpuRes[0,:], gpuRes[1,:]/gpuRes[0,:], label='Nvidia V100')
plt.plot(gpuRes[0,:], np.ones(len(gpuRes[0,:]))*averageCPUPerf, label='x4 Cores of Xeon Gold 6428')
plt.yscale('log')
plt.title('Comparing CPU and GPU Performance')
plt.legend()
plt.xlabel("Number of Particles at Once")
plt.ylabel("Time Taken Per Particle (s)")
plt.show()
```

```python
bestEfficiencyCPU = np.min(cpuRes[1,:]/cpuRes[0,:])
bestEfficiencyGPU = np.min(gpuRes[1,:]/gpuRes[0,:])
```
